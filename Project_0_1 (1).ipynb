##Pie Plot

A pie plot is a proportional representation of the numerical data in a column.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
df = pd.read_csv("/content/laptop price.csv")
df.groupby(['brand']).sum().plot( kind='pie', y='Unnamed: 0.1',autopct='%1.0f%%')
plt.show()
```

##Heat Maps

Heat Maps is a type of plot which is necessary when we need to find the dependent variables. One of the best way to find the relationship between the features can be done using heat maps.

```python
import seaborn as sns
sns.heatmap(df.corr(numeric_only="price"),cmap='BrBG',annot=True)
```
